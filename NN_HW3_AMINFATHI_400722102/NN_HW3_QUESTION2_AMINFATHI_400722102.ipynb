```python
def dist(a,b):
    out = 0
    for i in range(len(a)):
        out = out +  np.power(a[i] - b[i],2)
        out2 = np.sqrt(out)
    return out2

    # این تابع برای محاسبه فاصله نوشته شده است ، متاسفانه 
```
